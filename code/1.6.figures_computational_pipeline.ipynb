## 1.6. figures computational pipeline

This notebook creates the (sub)figures that are used in figure 1h of our manuscript (computational pipeline). This is probably a less interesting notebook to run, but for the sake of transparency I still wanted to make it available. 

```python
### loading python libraries

# standard libraries
import numpy as np
import pandas as pd

# single cell libraries
import anndata as ad
import scanpy as sc
sc.settings.verbosity = 0 

# plotting libraries
from matplotlib import pyplot as plt
plt.rcParams['pdf.fonttype'] = 42 #for saving PDF with changeable text
plt.rcParams['ps.fonttype'] = 42 #for saving PDF with changeable text

# other 
import warnings
warnings.filterwarnings('ignore') #ignore filter warnings
import sklearn as sk #used for L2 normalization
import sklearn.preprocessing #used for L2 normalization
```

### Load data

Here, we load the filtered HSPC dataset and the control subset(as preprocessed in notebook 1.1 and 1.3).

```python
### load Anndata object with the control subset 
adata_ctrl_path = "../data/count_matrices/control_filtered.h5ad"
adata_ctrl = sc.read_h5ad(adata_ctrl_path)
```

```python
### load Anndata object with the HSPC dataset (all four subsets incl.)
adata_path = "../data/count_matrices/all_filtered.h5ad"
adata = sc.read_h5ad(adata_path)
```

### Plot UMAPs of control subset

Here, we plot the UMAPs for the control subset (first part of the pipeline).

### Plot "raw" version of HSPC dataset
Lastly, we want a plot with the complete HSPC dataset, but without any batch correction (opposite to what is done in notebook 1.3). 

```python
### subset HSPC dataset with HVGs

# load HVGs
hvgs_path  = "../results/cluster_annotation/HVGs_top2000_all_timepoints_merged.csv"
hvgs = pd.read_csv(hvgs_path)
hvgs = np.array(hvgs).flatten()

# subset dataset
adata_hvgs = adata[:, hvgs].copy()
```

```python
### L2 normalize dataset
adata_hvgs.X = sk.preprocessing.normalize(adata_hvgs.X, norm='l2')
```

```python
### compute PCA  
sc.set_figure_params(figsize = (15,3))
sc.tl.pca(adata_hvgs, n_comps=50)
sc.pl.pca_variance_ratio(adata_hvgs, n_pcs=50)
```

```python
### calculate NN
sc.pp.neighbors(adata_hvgs, n_neighbors=10, n_pcs=15) 
```

```python
### calculate UMAP 
sc.tl.umap(adata_hvgs)
```

```python
### plot HSPC dataset without batch correction, colored for time
fig, ax = plt.subplots(1,1,figsize=(3,3))
ax.axis('off')
sc.pl.umap(adata_hvgs, size=15, ax=ax, color="time", title="", show=False, add_outline=True)
ax.get_legend().remove()
plt.show()

# save figure
save_path = "../figures/1.6.computational_pipeline_all_timepoints_without_batch_correction.png"
fig.savefig(save_path, bbox_inches='tight', format='png', dpi=300)
```

```python
### plot HSPC dataset without batch correction, colored for time (PCA)
pc_components = []

variance_dict = {0:"PC1\n {:0.3f}".format(adata_hvgs.uns['pca']['variance_ratio'][0]),
                 4:"PC2\n {:0.3f}".format(adata_hvgs.uns['pca']['variance_ratio'][1]),
                 8:"PC3\n {:0.3f}".format(adata_hvgs.uns['pca']['variance_ratio'][2]),}

fig, ax = plt.subplots(3, 3, figsize=(9,6), gridspec_kw={'wspace':0.05, 'hspace':0.05})

ax = ax.ravel()

for idx, components in enumerate([None,'2,1','3,1','1,2',None,'3,2','1,3','2,3', None]):
    if components == None:
        #ax[idx].axis('off')
        ax[idx].text(0.5, 0.5, variance_dict[idx], fontsize=12, horizontalalignment='center', verticalalignment='center')
        ax[idx].set_xlabel('')
        ax[idx].set_ylabel('')
        ax[idx].set_xticks([])
        ax[idx].set_yticks([])
        pass
    else:
        sc.pl.pca(adata_hvgs, size=15, ax=ax[idx], components=[components],
                  color="time", title="", show=False, add_outline=True)
        ax[idx].set_xlabel('')
        ax[idx].set_ylabel('')
        ax[idx].get_legend().remove()
plt.show()

### save figure
save_path = "../figures/1.6.PCA_pair-plot_HSPC-all_timepoints_without_batch_correction.png"
fig.savefig(save_path, bbox_inches='tight', format='png', dpi=300)
```
